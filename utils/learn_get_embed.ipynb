# 三, 加载Glove(txt形式)

```python
lines = []
with open('glove.6B.100d.txt', 'r') as f:
    lines = f.readlines()
```

```python
len(lines)
```

```python
dct = {}
for line in lines:
    data = line.strip().split()
    word = data[0]
    data.pop(0)
    dct[word] = np.array(data)
```
